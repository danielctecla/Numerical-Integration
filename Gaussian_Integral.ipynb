### 2. Problema
La integral de:
$$\int_{-6}^b \frac{1}{\sqrt{2π}}e^{-\frac{1}{2}z^2}dz$$

```python
f = lambda x: 1/smp.sqrt(2 * smp.pi) * smp.exp(-0.5 * (x**2))
```

```python
b = input("Ingresa b: ")
b = float(b)
```

```python
init = -6
increment = 0.01
i = 0

temp = init
result = []
while temp < b:
    temp = round(temp + increment, 2)
    result.append(quad(f,-6,temp)[0])
    print("b{:5}".format(i),end = "")
    print(" = {:.2f}".format(temp) ," I = {:7}".format(result[-1]))
    i += 1

```
