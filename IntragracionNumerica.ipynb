### 1. Problema
La integral de:
$$\int_a^b\frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{(\frac{x-\mu}{
  \sigma})}{2}}dx$$

```python
mu_value = float(input("Ingrese el valor de mu: "))
sigma_value = float(input("Ingrese el valor de sigma: "))
a_limit = float(input("Ingrese el valor de a: "))
b_limit = float(input("Ingrese el valor de b: "))
```

```python
# Definir la expresión a integrar
f_ = lambda x: 1/(sigma_value*smp.sqrt(2*smp.pi))*smp.exp(-(1/2)*((x-mu_value)**2/sigma_value**2))
```

```python
print("La aproximación de la integral es: ",quad(f_,a_limit,b_limit)[0])
```

### 2. Problema
La integral de:
$$\int_{-6}^b \frac{1}{\sqrt{2π}}e^{-\frac{1}{2}z^2}dz$$

```python

f = lambda x: 1/smp.sqrt(2 * smp.pi) * smp.exp(-0.5 * (x**2))
```

```python
b = input("Ingresa b: ")
b = float(b)
```

```python
init = -6
increment = 0.01
i = 0

temp = init
result = []
while temp < b:
    temp = round(temp + increment, 2)
    result.append(quad(f,-6,temp)[0])
    #print("b[",i,"] =",temp,"  I =",result[-1])
    print("b{:5}".format(i),end = "")
    print(" = {:.2f}".format(temp) ," I = {:10}".format(result[-1]))
    i += 1

```
