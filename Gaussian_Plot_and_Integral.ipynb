### 1. Problema
La integral de:
$$\int_a^b\frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{(\frac{x-\mu}{
  \sigma})^2}{2}}dx$$

```python
mu_value = float(input("Ingrese el valor de mu: "))
sigma_value = float(input("Ingrese el valor de sigma: "))
a_limit = float(input("Ingrese el valor de a: "))
b_limit = float(input("Ingrese el valor de b: "))
```

```python
# Definir la expresión a integrar
f_ = lambda x: 1/(sigma_value*smp.sqrt(2*smp.pi))*smp.exp(-(1/2)*((x-mu_value)**2/sigma_value**2))
f_numeric = smp.lambdify(smp.symbols('x'), f_(smp.symbols('x')), 'numpy')
x_values = np.linspace(mu_value - 4*sigma_value, mu_value + 4*sigma_value, 1000)
plt.plot(x_values, f_numeric(x_values), label=r'$\frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}$')
plt.fill_between(x_values,f_numeric(x_values),where = [(x > a_limit) and (x < b_limit) for x in x_values],color = 'green',alpha = 0.3)
plt.title('Gaussian Distribution')
plt.axhline(color='black')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
print("La aproximación de la integral es: ",quad(f_,a_limit,b_limit)[0])
```
